```python
import requests
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from bs4 import BeautifulSoup
import pandas as pd
from io import StringIO

def get_data_with_selenium():
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument('--headless')
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--disable-dev-shm-usage')
    chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3")

    driver = webdriver.Chrome(options=chrome_options)

    try:
        url = "https://www.spaceweatherlive.com/en/solar-activity/top-50-solar-flares"
        driver.get(url)

        WebDriverWait(driver, 60).until(
            EC.presence_of_element_located((By.XPATH, "//table[contains(@class, 'table-striped')]"))
        )

        page_source = driver.page_source
        return page_source

    except TimeoutException:
        print("Loading took too much time!")
        return None

    finally:
        driver.quit()

def get_data_with_requests():
    url = "https://www.spaceweatherlive.com/en/solar-activity/top-50-solar-flares"
    response = requests.get(url)
    return response.text
def parse_data(html_content):
    soup = BeautifulSoup(html_content, "html.parser")
    table = soup.find("table", {"class": "table table-striped"})

    if table:
        # Print  the prettified table HTML
        print("Prettified Table Content:")
        print(table.prettify())
        return True
    else:
        print("Table not found on the page.")
        return False

# Main execution
html_content = get_data_with_selenium()

if html_content is None:
    print("Selenium failed. Trying with requests...")
    html_content = get_data_with_requests()

if html_content:
    parse_data(html_content)
else:
    print("Failed to retrieve the webpage.")

```

```python
import pandas as pd
import numpy as np

def tidy_solar_flare_data(df):
    # Convert rank to integer
    df['rank'] = pd.to_numeric(df['rank'], errors='coerce').astype('Int64')

    # Ensure region is of type Int64 (allows for NaN values)
    df['region'] = pd.to_numeric(df['region'], errors='coerce').astype('Int64')

    # Handle x_class conversion more robustly
    def convert_x_class(x):
        x = str(x).replace('+', '')  # Remove '+' if present
        try:
            return float(x[1:])  # Remove 'X' and convert to float
        except ValueError:
            return np.nan  # Return NaN if conversion fails

    df['x_class'] = df['x_class'].apply(convert_x_class)

    # Convert date to datetime
    df['date'] = pd.to_datetime(df['date'], errors='coerce')

    # Combine date with time columns to create datetime columns
    time_columns = ['start_time', 'max_time', 'end_time']
    for col in time_columns:
        df[f'{col}_datetime'] = pd.to_datetime(df['date'].astype(str) + ' ' + df[col], errors='coerce')

    return df

# Apply the tidying function
df_tidied = tidy_solar_flare_data(df)

# Display the first few rows of the tidied data
print(df_tidied.head().to_string())

# Save the tidied data to a CSV file
df_tidied.to_csv("top_50_solar_flares_tidied.csv", index=False)
print("Tidied data saved to 'top_50_solar_flares_tidied.csv'")
```

Step 3: Scraping the Nasa Data

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

def clean_html(raw_html):
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr, '', raw_html)
    return cleantext.strip()

def scrape_nasa_data():
    url = "http://cdaw.gsfc.nasa.gov/CME_list/radio/waves_type2.html"
    response = requests.get(url)

    if response.status_code != 200:
        print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
        return None

    soup = BeautifulSoup(response.content, 'html.parser')
    pre_content = soup.find('pre')

    if not pre_content:
        print("Could not find the <pre> tag containing the data.")
        return None

    # Clean the content
    clean_content = clean_html(str(pre_content))

    # Split into lines and remove header
    lines = clean_content.split('\n')[7:]  # Skip the first 7 lines to remove all headers

    data = []
    for line in lines:
        if line.strip() and not line.startswith('='):  # Skip empty lines and separators
            parts = line.split()
            if len(parts) >= 14:  # Ensure we have all expected columns
                data.append(parts[:14])  # Take only the first 14 columns

    columns = [
        'start_date', 'start_time', 'end_date', 'end_time',
        'start_frequency', 'end_frequency', 'solar_source_location',
        'noaa_active_region', 'xray_flare_importance', 'cme_date',
        'cme_time', 'cme_angle', 'cme_width', 'cme_speed'
    ]

    df = pd.DataFrame(data, columns=columns)

    # Convert date columns to datetime
    df['start_date'] = pd.to_datetime(df['start_date'] + ' ' + df['start_time'], format='%Y/%m/%d %H:%M', errors='coerce')
    df['end_date'] = pd.to_datetime(df['start_date'].dt.strftime('%Y/') + df['end_date'] + ' ' + df['end_time'], format='%Y/%m/%d %H:%M', errors='coerce')
    df['cme_date'] = pd.to_datetime(df['start_date'].dt.strftime('%Y/') + df['cme_date'] + ' ' + df['cme_time'], format='%Y/%m/%d %H:%M', errors='coerce')

    # Convert numeric columns
    numeric_cols = ['start_frequency', 'end_frequency', 'cme_angle', 'cme_width', 'cme_speed']
    for col in numeric_cols:
        df[col] = pd.to_numeric(df[col].replace('?', ''), errors='coerce')

    # Clean up special values
    df = df.replace({'----': pd.NA, '-----': pd.NA, '--:--': pd.NA})

    return df

nasa_data = scrape_nasa_data()
if nasa_data is not None and not nasa_data.empty:
    print(nasa_data.head())
    nasa_data.to_csv('nasa_solar_flare_data.csv', index=False)
    print("Data saved to 'nasa_solar_flare_data.csv'")
    print(f"Total number of records: {len(nasa_data)}")
else:
    print("Failed to scrape NASA data or the resulting DataFrame is empty.")
```

```python
import pandas as pd
import numpy as np

def tidy_nasa_data(df):
    df = df.copy()

    # 1. Recode missing entries as NaN
    df = df.replace({
        '----': np.nan, '-----': np.nan, '--': np.nan, '???': np.nan,
        '(7)': np.nan, '(8)': np.nan, '(9)': np.nan, '(2)': np.nan, '(4)': np.nan, '(11)': np.nan
    })

    # 2. Handle CPA column (cme_angle)
    df['is_halo'] = df['cme_angle'].fillna('').astype(str).str.lower() == 'halo'
    df['cme_angle'] = pd.to_numeric(df['cme_angle'].replace('Halo', np.nan), errors='coerce')

    # 3. Handle width column
    df['width_lower_bound'] = df['cme_width'].fillna('').astype(str).str.startswith('>')
    df['cme_width'] = pd.to_numeric(df['cme_width'].fillna('').astype(str).str.replace('>', ''), errors='coerce')

    # 4. Combine date and time columns
    for prefix in ['start', 'end', 'cme']:
        date_col = f'{prefix}_date'
        time_col = f'{prefix}_time'
        if date_col in df.columns and time_col in df.columns:
            df[f'{prefix}_datetime'] = df.apply(
                lambda row: pd.to_datetime(f"{row[date_col].strftime('%Y-%m-%d')} {row[time_col]}",
                                           format='%Y-%m-%d %H:%M',
                                           errors='coerce') if pd.notnull(row[date_col]) else pd.NaT,
                axis=1
            )
            df = df.drop(columns=[date_col, time_col])

    # Convert numeric columns
    numeric_cols = ['start_frequency', 'end_frequency', 'cme_speed', 'flare_region']
    for col in numeric_cols:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors='coerce')

    # Rename columns
    rename_dict = {
        'solar_source_location': 'flare_location',
        'noaa_active_region': 'flare_region',
        'xray_flare_importance': 'importance'
    }
    df = df.rename(columns={k: v for k, v in rename_dict.items() if k in df.columns})

    # Clean up flare_location
    df['flare_location'] = df['flare_location'].replace({'------': np.nan, 'BACK': np.nan})
    df['flare_location'] = df['flare_location'].where(~df['flare_location'].str.contains('90', na=False), np.nan)

    # Define the desired column order
    desired_columns = [
        'start_datetime', 'end_datetime', 'start_frequency', 'end_frequency',
        'flare_location', 'flare_region', 'importance',
        'cme_datetime', 'cme_angle', 'cme_width', 'cme_speed',
        'is_halo', 'width_lower_bound'
    ]

    # Only include columns that exist in the DataFrame
    column_order = [col for col in desired_columns if col in df.columns]

    return df[column_order]

# Assuming nasa_data is your DataFrame
tidied_nasa_data = tidy_nasa_data(nasa_data)

# Print the first few rows of the tidied data
print(tidied_nasa_data.head().to_string())

# Print column information
print("\nColumn Info:")
print(tidied_nasa_data.info())

# Print summary statistics
print("\nSummary Statistics:")
print(tidied_nasa_data.describe())

# Save the tidied data to a new CSV file
tidied_nasa_data.to_csv('tidied_nasa_solar_flare_data.csv', index=False)
print("\nTidied data saved to 'tidied_nasa_solar_flare_data.csv'")

# Print unique values in 'flare_location' column
print("\nUnique values in 'flare_location' column:")
print(tidied_nasa_data['flare_location'].unique())
```

```python
import pandas as pd
import re

# Load the NASA data
nasa_data = pd.read_csv('tidied_nasa_solar_flare_data.csv')

# Load the SpaceWeatherLive data
spaceweather_data = pd.read_csv('top_50_solar_flares.csv')

print("SpaceWeatherLive Data Columns:")
print(spaceweather_data.columns)

def flare_class_to_numeric(flare_class):
    if pd.isna(flare_class):
        return 0
    match = re.match(r'([ABCMX])(\d+(\.\d+)?)', flare_class)
    if match:
        letter, number = match.groups()[:2]
        multiplier = {'A': 1e-8, 'B': 1e-7, 'C': 1e-6, 'M': 1e-5, 'X': 1e-4}
        return multiplier[letter] * float(number)
    return 0

# Convert flare classification to numeric values for NASA data
nasa_data['flare_numeric'] = nasa_data['importance'].apply(flare_class_to_numeric)

# Sort NASA data by flare_numeric in descending order and select top 50
top_50_nasa = nasa_data.sort_values('flare_numeric', ascending=False).head(50)

# Display the top 10 flares from NASA data
print("\nTop 10 flares from NASA data:")
print(top_50_nasa[['start_datetime', 'flare_location', 'importance', 'flare_numeric']].head(10))

# Use the correct column names for SpaceWeatherLive data
correct_columns = ['x_class', 'region', 'date', 'start_time']

# Display the top 10 flares from SpaceWeatherLive data using correct column names
print("\nTop 10 flares from SpaceWeatherLive data:")
print(spaceweather_data[correct_columns].head(10))

# Compare the number of matching flares
nasa_classes = set(top_50_nasa['importance'])
spaceweather_classes = set(spaceweather_data['x_class'])
matching_classes = nasa_classes.intersection(spaceweather_classes)

print(f"\nNumber of matching flare classes: {len(matching_classes)}")
print(f"Matching classes: {sorted(matching_classes)}")
```

Question 2: A function to find the best matching row in the NASA Data for each of the top 50 solar flares in SpaceWeatherLive data.

```python
import pandas as pd
from datetime import timedelta

# Load the NASA data
nasa_data = pd.read_csv('tidied_nasa_solar_flare_data.csv')

# Load the SpaceWeatherLive data
spaceweather_data = pd.read_csv('top_50_solar_flares.csv')

print("NASA Data Columns:")
print(nasa_data.columns)
print("\nSpaceWeatherLive Data Columns:")
print(spaceweather_data.columns)

# Check data types
print("\nNASA Data Types:")
print(nasa_data.dtypes)
print("\nSpaceWeatherLive Data Types:")
print(spaceweather_data.dtypes)

# Convert datetime columns
nasa_data['start_datetime'] = pd.to_datetime(nasa_data['start_datetime'], errors='coerce')
spaceweather_data['date'] = pd.to_datetime(spaceweather_data['date'], errors='coerce')

def flare_class_to_numeric(flare_class):
    if pd.isna(flare_class):
        return 0
    match = re.match(r'([ABCMX])(\d+(\.\d+)?)', str(flare_class))
    if match:
        letter, number = match.groups()[:2]
        multiplier = {'A': 1e-8, 'B': 1e-7, 'C': 1e-6, 'M': 1e-5, 'X': 1e-4}
        return multiplier[letter] * float(number)
    return 0

# Convert flare classification to numeric values for NASA data
nasa_data['flare_numeric'] = nasa_data['importance'].apply(flare_class_to_numeric)

def find_best_matches(spaceweather_data, nasa_data):
    matches = []

    for _, sw_flare in spaceweather_data.iterrows():
        sw_date = pd.to_datetime(sw_flare['date'])
        sw_class = sw_flare['x_class']
        sw_region = sw_flare['region']

        # Filter NASA data for the same date (allowing 1 day difference)
        date_matches = nasa_data[
            (nasa_data['start_datetime'].dt.date >= sw_date.date() - timedelta(days=1)) &
            (nasa_data['start_datetime'].dt.date <= sw_date.date() + timedelta(days=1))
        ]

        if date_matches.empty:
            matches.append(None)
            continue

        # Score each potential match
        date_matches['score'] = 0

        # Score based on flare class match
        date_matches.loc[date_matches['importance'] == sw_class, 'score'] += 3

        # Score based on region match (if available)
        if pd.notna(sw_region):
            date_matches.loc[date_matches['flare_region'] == sw_region, 'score'] += 2

        # Score based on start time proximity
        sw_start_time = pd.to_datetime(sw_flare['date'].strftime('%Y-%m-%d') + ' ' + sw_flare['start_time'])
        date_matches['time_diff'] = abs(date_matches['start_datetime'] - sw_start_time)
        date_matches.loc[date_matches['time_diff'] <= timedelta(hours=1), 'score'] += 1

        # Get the best match
        best_match = date_matches.loc[date_matches['score'].idxmax()]
        matches.append(best_match.name if best_match['score'] > 0 else None)

    return matches

# Apply the function
best_matches = find_best_matches(spaceweather_data, nasa_data)

# Add rank to NASA dataset
nasa_data['spaceweather_rank'] = pd.Series(dtype='Int64')
for rank, match_index in enumerate(best_matches, 1):
    if match_index is not None:
        nasa_data.loc[match_index, 'spaceweather_rank'] = rank

# Display some results
print("\nSample of NASA data with SpaceWeatherLive ranks:")
print(nasa_data[nasa_data['spaceweather_rank'].notna()].sort_values('spaceweather_rank').head(10))

print("\nNumber of matched flares:", nasa_data['spaceweather_rank'].notna().sum())

# Display the top 10 flares from NASA data
print("\nTop 10 flares from NASA data:")
print(nasa_data.sort_values('flare_numeric', ascending=False)[['start_datetime', 'flare_location', 'importance', 'flare_numeric', 'spaceweather_rank']].head(10))

# Display the top 10 flares from SpaceWeatherLive data
print("\nTop 10 flares from SpaceWeatherLive data:")
print(spaceweather_data[['date', 'start_time', 'x_class', 'region']].head(10))

# Compare the number of matching flares
nasa_classes = set(nasa_data['importance'])
spaceweather_classes = set(spaceweather_data['x_class'])
matching_classes = nasa_classes.intersection(spaceweather_classes)

print(f"\nNumber of matching flare classes: {len(matching_classes)}")
print(f"Matching classes: {sorted(matching_classes)}")
```
